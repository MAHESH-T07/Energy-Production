we can see outliers in ambient pressure to we should drop those outliers using quartile concept

```python
q1 = np.quantile(energy_data2['amb_pressure'],0.25)
q3 = np.quantile(energy_data2['amb_pressure'],0.75)
```

```python
IQR = q3-q1
```

```python
lower_bound
```

```python
upper_bound
```

```python
# considering only the data points which are not outliers
energy_data3=energy_data2[(energy_data2['amb_pressure']>lower_bound) & (energy_data2['amb_pressure']<upper_bound)]
energy_data3.head()
```

```python
print('df length before trating outliers:',len(energy_data2))
print('df lenght after treating outliers:',len(energy_data3))
print('ouliers length:',len(energy_data2)-len(energy_data3))
```

### Use stats model to decide whether we have to keep tempareture column for model building or not to take

## VIF

```python
temp_model = sfa.ols('temperature~exhaust_vacuum+amb_pressure+humidity',data=energy_data4).fit()
exh_model = sfa.ols('exhaust_vacuum~temperature+amb_pressure+humidity',data=energy_data4).fit()
amb_model = sfa.ols('amb_pressure~exhaust_vacuum+temperature+humidity',data=energy_data4).fit()
hum_model = sfa.ols('humidity~amb_pressure+exhaust_vacuum+temperature',data=energy_data4).fit()
```

```python
temp_vif=1/(1-temp_model.rsquared)
exh_vif=1/(1-exh_model.rsquared)
amb_vif=1/(1-amb_model.rsquared)
hum_vif=1/(1-hum_model.rsquared)
```

```python
pd.DataFrame(index=['temp','vacuum','pressure','humidity'],columns=['VIF','Rsuared'],data=[[temp_vif,temp_model.rsquared],[exh_vif,exh_model.rsquared],[amb_vif,amb_model.rsquared],[hum_vif,hum_vif]])
```

Considering tempareture for the model building because, the hypothesis says it good to consider the tempareture column because the
p value < 0.05 and vif value should less 5 to consider but it ok we go with tempareture columns for model building.

# Xgboost Model

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
```

```python
xgb_model = XGBRegressor()
```

```python
param = {
    'n_estimators' : [13,18,20,12],
    'max_depth' : [8,12,16,20],
    'learning_rate':[0.11,0.12,0.13,0.1],
    'subsample':[0.5,0.7,0.8,0.9]
}
```

```python
grid = GridSearchCV(estimator=xgb_model,param_grid=param,cv=10,verbose=10)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_score_
```

```python
grid.best_params_
```

```python
xgb_model = grid.best_estimator_
```

```python
xtr_pred = xgb_model.predict(X_train)
xts_pred = xgb_model.predict(X_test)
```

### metrics

```python
xtr_mse = mean_squared_error(xtr_pred,y_train)
xtr_rmse = np.sqrt(xtr_mse)
xtr_mse,xtr_rmse
```

```python
xts_mse = mean_squared_error(xts_pred,y_test)
xts_rmse = np.sqrt(xts_mse)
xts_mse,xts_rmse
```

```python
xtr_r2score=r2_score(xtr_pred,y_train)
xtr_r2score
```

```python
xts_r2score=r2_score(xts_pred,y_test)
xts_r2score
```

### cross validation test

```python
from sklearn.model_selection import cross_val_score
```

```python
scores = cross_val_score(xgb_model,X_train,y_train,cv=5,scoring='r2')
print(f"cross validation scores {scores}")
print(f"mean of cross validation {scores.mean()}")
```

The train R score (0.98) is higher than the test R score, but the difference isn't significant (0.04), suggesting the model is niether overfit nor underfit.

The cross-validation scores are consistent and very close to the test R score, showing that the model generalizes well across different data splits

## Importing stacking model to use both model

```python
from sklearn.ensemble import StackingRegressor
```

```python
base_model = [
    ('linear', lin_model),
    ('xgb' , xgb_model)
]
```

```python
meta_model = XGBRegressor(learning_rate= 0.13, max_depth= 7, n_estimators= 20, subsample= 0.9)
```

```python
stacking_model = StackingRegressor(estimators=base_model,final_estimator=meta_model)
```

```python
stacking_model.fit(X_train,y_train)
```

```python
stacktr_pred = stacking_model.predict(X_train)
stackts_pred = stacking_model.predict(X_test)
```

```python
stc_residuals = y_test - stackts_pred.reshape(-1,1)
```

```python
sns.histplot(stc_residuals,kde=True)
plt.show()
```

The residuals are normally distributed

```python
plt.scatter(stackts_pred,stc_residuals)
plt.axhline(0, color='red', linestyle='--')
plt.show()
```

The residuals distributed constantly in the range of -5 to 5
